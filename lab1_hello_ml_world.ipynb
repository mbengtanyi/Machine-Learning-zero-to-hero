```
import tensorflow as tf
import numpy as np
from tensorflow import keras
#keras is the framework for defining a neural network as a set of sequential layers
```

```
#Creating the simple Neural Network(NN) that has 1 layer with 1 neuron with 1 as input shape
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```
#Loss function measures guessed answers against known correct answers. Also the measure of how good or bad the prediction is
#Optimizer function makes another guess based on the prev loss funtion guess and tries to minimize the loss and get closer to the result.
#It repeats for the number of epochs
#sgd = stochastic gradient descent
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

#Feed in data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```
#Training the neural network using fit method just like when you're in the process of getting a regplot in datascience
model.fit(xs, ys, epochs = 500)
```

```
print(model.predict([10.0]))
```
